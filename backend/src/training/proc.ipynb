```python
import pandas as pd

import pandas as pd
df = pd.read_csv('products_data_for_train.csv')
print(f"Всего товаров: {len(df)}")
print(f"Уникальных категорий: {df['category'].nunique()}")
print(f"\nТоваров на категорию (среднее): {len(df) / df['category'].nunique():.1f}")
print("\n" + "="*60)
print("ТОП-20 КАТЕГОРИЙ (по количеству товаров):")
print("="*60)
category_counts = df['category'].value_counts()
print(category_counts.head(20))
print("\n" + "="*60)
print("КАТЕГОРИИ С МАЛЫМ КОЛИЧЕСТВОМ ТОВАРОВ (< 5):")
print("="*60)
small_categories = category_counts[category_counts < 5]
print(f"Количество: {len(small_categories)}")
print(small_categories)
```
